# Python program for implementation of Quick Sort
Algorithm: quick sort

- Best_case  : O(nlogn) when paritioning divides into two equal halves.
- worst_case : O(n)2 when list is already sorted, program fails to divide the sequence into smaller problems.
- Avg_case   : O(nlogn)
    
- Auxiliary Space: O(1) #Quick Sort in its general form is an in-place sort.
(i.e. it doesnâ€™t require any extra storage) whereas merge sort requires O(N) extra storage, 
N denoting the array size which may be quite expensive.

- Sorting In Place: Yes,  it does not require extra space.


```python
# This function takes last element as pivot, places
# the pivot element at its correct position in sorted
# array, and places all smaller (smaller than pivot)
# to left of pivot and all greater elements to right of pivot
```

```python
quickcounter = 0
def partition(arr,low,high):
    
    pIndex = low         # index of smaller element
    pivot = arr[high]    # pivot
    global quickcounter
    quickcounter += 1
    
    print("Pass:", quickcounter, "With Pivot: %d"%pivot)
    print("Partion under process %s" %str(arr[low:high+1]))
    for j in range(low , high): #pivot is excluded loop runs from 0 to 5 [64, 34, 25, 12, 22, 11, 9] means only till 11
        
        print("Interation: ", j)
        # If current element is smaller than or
        # equal to pivot
        if   arr[j] <= pivot:
         
            # increment index of smaller element
            #i = i+1
            arr[pIndex],arr[j] = arr[j],arr[pIndex]
            pIndex +=1
            print("Inside for loop->swapping happening->", arr[pIndex],arr[j])
            
    arr[pIndex],arr[high] = arr[high],arr[pIndex]
    print("outside for loop->swapping happening->", arr[pIndex],arr[high])
    print("Array :%s" %str(arr))
    print("Pivot placed at position :%s" %pIndex)
    print()
    return ( pIndex )
    
 
# The main function that implements QuickSort
# arr[] --> Array to be sorted,
# low  --> Starting index,
# high  --> Ending index
 
# Function to do Quick sort
def quickSort(arr,low,high):
    if low < high:
 
        # pi is partitioning index, arr[p] is now
        # at right place
        pi = partition(arr,low,high)
 
        # Separately sort elements before
        # partition and after partition
        quickSort(arr, low, pi-1)
        quickSort(arr, pi+1, high)
 
# Driver code to test above
arr = [64, 34, 25, 12, 22, 11, 9]
print("Input: %s" %str(arr))
n = len(arr)
quickSort(arr,0,n-1)
print ("Sorted array is:")
for i in range(n):
    print ("%d" %arr[i]),
```
